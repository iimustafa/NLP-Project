```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from datasets import load_dataset
# You'll need to download these NLTK data if you haven't already
# nltk.download('punkt')

def load_corpus_from_hf_dataset(dataset_name="Bluckr/assistant_spanish_pofi_v1", text_column="text"):
    """
    Loads text data from a Hugging Face dataset, preprocesses it,
    and returns a list of tokenized sentences.

    Args:
        dataset_name (str): The name of the Hugging Face dataset.
        text_column (str): The name of the column in the dataset containing the text.

    Returns:
        list: A list of lists, where each inner list is a tokenized sentence.
    """
    sentences = []

    # 1. Load the dataset from Hugging Face Hub
    # You might want to specify a split, e.g., ds = load_dataset(dataset_name, split='train')
    ds = load_dataset(dataset_name)

    # Hugging Face datasets often come with multiple splits (train, validation, test)
    # You'll need to decide which split(s) you want to use.
    # For simplicity, let's iterate through all available splits.
    # If your dataset only has 'train', you can directly access ds['train']
    for split in ds.keys():
        print(f"Processing split: {split}")
        for example in ds[split]:
            text = example[text_column]

            # Ensure text is string and convert to lowercase
            if isinstance(text, str):
                text = text.lower()
            else:
                # Handle cases where the 'text_column' might not contain strings
                # For example, if it's None or another data type, skip it.
                continue

            # Simple sentence & word tokenization
            for sent in sent_tokenize(text, language='spanish'): # Specify language for Spanish tokenization
                tokens = [w for w in word_tokenize(sent, language='spanish') if w.isalpha()]
                if tokens:
                    sentences.append(tokens)
    return sentences

# --- How to use it ---
# Make sure you have NLTK punkt tokenizer data downloaded for Spanish
try:
    nltk.data.find('tokenizers/punkt/spanish.pickle')
except LookupError:
    nltk.download('punkt') # This will download punkt for all languages, or specific ones.

# Customize with your dataset information
# Assuming 'Bluckr/assistant_spanish_pofi_v1' has a 'text' column.
# You can inspect your dataset structure after loading: ds.features
# If the text is in a different column, change text_column accordingly.
corpus = load_corpus_from_hf_dataset(dataset_name="Bluckr/assistant_spanish_pofi_v1", text_column="text")

# Now 'corpus' contains your tokenized sentences, ready for Word2Vec
print(f"Total sentences in corpus: {len(corpus)}")
print(f"First 5 sentences: {corpus[:5]}")

# Example Word2Vec training
from gensim.models import Word2Vec

if corpus: # Only train if there are sentences
    model = Word2Vec(sentences=corpus, vector_size=100, window=5, min_count=1, workers=4)
    print("Word2Vec model trained successfully!")
    # You can then use the model:
    # print(model.wv.most_similar('hola'))
else:
    print("Corpus is empty. Cannot train Word2Vec model.")
```

#Clustering your vocabulary

```python
from sklearn.cluster import KMeans

# take top-N frequent words and cluster them
top_n = 200
vectors = [model.wv[w] for w in model.wv.index_to_key[:top_n]]
kmeans = KMeans(n_clusters=8, random_state=0).fit(vectors)

clusters = {}
for w, label in zip(model.wv.index_to_key[:top_n], kmeans.labels_):
    clusters.setdefault(label, []).append(w)

for label, words in clusters.items():
    print(f"Cluster {label}:", words[:10])
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from datasets import load_dataset
from gensim.models import Word2Vec
import gradio as gr

# Ensure NLTK punkt tokenizer data for Spanish is downloaded
try:
    nltk.data.find('tokenizers/punkt/spanish.pickle')
except LookupError:
    nltk.download('punkt')

# --- 1. Load and preprocess the dataset (as per your previous setup) ---
def load_corpus_from_hf_dataset(dataset_name="Bluckr/assistant_spanish_pofi_v1", text_column="text"):
    sentences = []
    ds = load_dataset(dataset_name)

    for split in ds.keys():
        print(f"Processing split: {split}")
        for example in ds[split]:
            text = example.get(text_column) # Use .get() for safer access

            if isinstance(text, str):
                text = text.lower()
                for sent in sent_tokenize(text, language='spanish'):
                    tokens = [w for w in word_tokenize(sent, language='spanish') if w.isalpha()]
                    if tokens:
                        sentences.append(tokens)
    return sentences

print("Loading and preprocessing corpus...")
corpus = load_corpus_from_hf_dataset(dataset_name="Bluckr/assistant_spanish_pofi_v1", text_column="text")
print(f"Corpus loaded with {len(corpus)} sentences.")

if not corpus:
    print("Corpus is empty. Cannot train Word2Vec models.")
else:
    # --- 2. Train Two Word2Vec Models (CBOW and Skip-gram) ---
    print("Training CBOW Word2Vec model...")
    # CBOW (Continuous Bag of Words) - default is sg=0
    cbow_model = Word2Vec(sentences=corpus, vector_size=100, window=5, min_count=5, workers=4, sg=0)
    print("CBOW model trained successfully.")

    print("Training Skip-gram Word2Vec model...")
    # Skip-gram - set sg=1
    skipgram_model = Word2Vec(sentences=corpus, vector_size=100, window=5, min_count=5, workers=4, sg=1)
    print("Skip-gram model trained successfully.")

    # --- Optional: Save models (good practice for larger models) ---
    # cbow_model.save("cbow_word2vec.model")
    # skipgram_model.save("skipgram_word2vec.model")

    # --- Get some example words for the UI hint ---
    example_word_cbow = None
    if cbow_model.wv:
        for word in ["hola", "gracias", "ayuda", "sistema"]:
            if word in cbow_model.wv.key_to_index:
                example_word_cbow = word
                break
    if not example_word_cbow and cbow_model.wv:
        # Fallback to a random word if common ones aren't found
        if cbow_model.wv.key_to_index:
            example_word_cbow = list(cbow_model.wv.key_to_index.keys())[0]
```

Build Gradio App

```python
# --- 3. Gradio App Setup ---

# Check if models were trained successfully
if not corpus:
    print("Cannot launch Gradio app: Corpus is empty. Models were not trained.")
else:
    # Define the prediction function for Gradio
    def get_similar_words(word: str, model_type: str) -> str:
        word = word.lower() # Convert input word to lowercase for consistency

        model_to_use = None
        if model_type == "CBOW":
            model_to_use = cbow_model
        elif model_type == "Skip-gram":
            model_to_use = skipgram_model
        else:
            return "Error: Invalid model type selected."

        if not model_to_use.wv:
            return "Error: Word2Vec model not loaded or trained correctly."

        if word not in model_to_use.wv.key_to_index:
            return f"Lo siento, la palabra '{word}' no se encontró en el vocabulario del modelo '{model_type}'. Por favor, intenta con otra palabra." \
                   f"\n(Sorry, the word '{word}' was not found in the vocabulary of the '{model_type}' model. Please try another word.)"

        try:
            # Get top 10 most similar words
            similar_words = model_to_use.wv.most_similar(word, topn=10)

            # Format the output
            output_str = f"Las 10 palabras más similares a '{word}' (modelo {model_type}):\n"
            for similar_word, similarity in similar_words:
                output_str += f"- {similar_word}: {similarity:.4f}\n"
            return output_str
        except Exception as e:
            return f"Ocurrió un error al buscar palabras similares: {e}"

    # Set an example input based on a word found in the vocabulary, if possible
    example_input = [
        [example_word_cbow or "hola", "CBOW"], # Use a dynamic example word
        ["gracias", "Skip-gram"]
    ]

    # Create the Gradio interface
    iface = gr.Interface(
        fn=get_similar_words,
        inputs=[
            gr.Textbox(
                label="Introduce una palabra en español (Enter a Spanish word)",
                placeholder=f"Ejemplo: {example_word_cbow or 'hola'}" # Provide an example
            ),
            gr.Radio(
                ["CBOW", "Skip-gram"],
                label="Selecciona el tipo de modelo (Select model type)",
                value="CBOW" # Default selection
            )
        ],
        outputs="text",
        title="Buscador de Palabras Similares (Word Similarity Finder)",
        description="Introduce una palabra en español para encontrar las 10 palabras más similares según dos modelos de Word2Vec (CBOW y Skip-gram) entrenados con el dataset Bluckr/assistant_spanish_pofi_v1."
                    "\n(Enter a Spanish word to find the 10 most similar words according to two Word2Vec models (CBOW and Skip-gram) trained on the Bluckr/assistant_spanish_pofi_v1 dataset.)",
        examples=example_input
    )

    # Launch the Gradio app
    print("\nLaunching Gradio app...")
    iface.launch(debug=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming cbow_model and skipgram_model are already trained and available
# from your previous code execution.

def plot_similarities_comparison(word, cbow_model, skipgram_model, topn=10):
    """
    Plots the top N most similar words and their similarities for CBOW and Skip-gram models.
    """
    word = word.lower()

    if word not in cbow_model.wv.key_to_index:
        print(f"Warning: '{word}' not in CBOW model vocabulary. Skipping plot.")
        cbow_similar = []
    else:
        cbow_similar = cbow_model.wv.most_similar(word, topn=topn)

    if word not in skipgram_model.wv.key_to_index:
        print(f"Warning: '{word}' not in Skip-gram model vocabulary. Skipping plot.")
        skipgram_similar = []
    else:
        skipgram_similar = skipgram_model.wv.most_similar(word, topn=topn)

    # --- Prepare data for plotting ---
    cbow_words = [item[0] for item in cbow_similar]
    cbow_scores = [item[1] for item in cbow_similar]

    skipgram_words = [item[0] for item in skipgram_similar]
    skipgram_scores = [item[1] for item in skipgram_similar]

    # Handle cases where one list might be shorter if a model has fewer similar words
    # or the input word was missing from one model
    max_len = max(len(cbow_words), len(skipgram_words))

    fig, axes = plt.subplots(1, 2, figsize=(16, 7), sharey=False) # Changed sharey to False for better scale comparison if needed
    fig.suptitle(f"Top {topn} palabras similares a '{word}' (Top {topn} similar words to '{word}')", fontsize=16)

    # Plot for CBOW
    if cbow_words:
        axes[0].barh(cbow_words, cbow_scores, color='skyblue')
        axes[0].set_title(f"Modelo CBOW", fontsize=14)
        axes[0].set_xlabel("Similitud (Similarity)", fontsize=12)
        axes[0].invert_yaxis() # Highest similarity at the top
        axes[0].set_xlim(0, 1) # Similarity scores are between 0 and 1
        for i, score in enumerate(cbow_scores):
            axes[0].text(score + 0.01, i, f'{score:.2f}', va='center') # Add score labels
    else:
        axes[0].set_title("Modelo CBOW", fontsize=14)
        axes[0].text(0.5, 0.5, "Datos no disponibles", transform=axes[0].transAxes, ha="center", va="center")


    # Plot for Skip-gram
    if skipgram_words:
        axes[1].barh(skipgram_words, skipgram_scores, color='lightcoral')
        axes[1].set_title(f"Modelo Skip-gram", fontsize=14)
        axes[1].set_xlabel("Similitud (Similarity)", fontsize=12)
        axes[1].invert_yaxis() # Highest similarity at the top
        axes[1].set_xlim(0, 1) # Similarity scores are between 0 and 1
        for i, score in enumerate(skipgram_scores):
            axes[1].text(score + 0.01, i, f'{score:.2f}', va='center') # Add score labels
    else:
        axes[1].set_title("Modelo Skip-gram", fontsize=14)
        axes[1].text(0.5, 0.5, "Datos no disponibles", transform=axes[1].transAxes, ha="center", va="center")

    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
    plt.show()

# --- Call the plotting function for 'hola' ---
plot_similarities_comparison('hola', cbow_model, skipgram_model)

# --- Call the plotting function for 'gracias' ---
plot_similarities_comparison('gracias', cbow_model, skipgram_model)
```

# ***Summary & Insights***

- **CBOW model** proved more effective at capturing functional, domain-specific semantic relationships crucial for an assistant-focused dataset. It consistently highlighted immediate, actionable contexts and entities.

- **Skip-gram model** performing well, leaned towards broader, more generalized conversational or abstract associations.



CBOW's Functional Insight:

- For "**hola**" (***hello***), CBOW directly connected to words like "**puedes**" (***you can***) and "**pofi**" (the assistant's name), indicating an understanding of common commands and direct interaction.

- For "**gracias**" (t***hank you***), CBOW linked to "**carolina**" (a likely named entity) and action-oriented terms like "enviando" (***sending***) and "**mensaje**" (***message***), suggesting a strong grasp of transactional gratitude.

- **Strength**: Excellent at discerning direct, local context and operational language.

Skip-gram's Broader Semantic View:

- For "**hola**," Skip-gram generated more general conversational words like "**estás**" (***you are***) and "**tal**" (***how***), reflecting general pleasantries.

- For "**gracias**," it associated with "**amistad**" (***friendship***) and "**incondicional**" (***unconditional***), pointing to wider emotional or social nuances.

- **Strength**: Better at learning abstract semantic similarities and capturing broader conceptual relationships.
